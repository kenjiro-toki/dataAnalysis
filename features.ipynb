```python
import opensmile
```

```python
import glob
files = sorted(glob.glob("/content/wav/*")) #sortしないと順番がメチャクチャになるので注意
print(files)
```

```python
import pandas as pd
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv01b, #eGeMAPSv01bという特徴量を使用する
    feature_level=opensmile.FeatureLevel.Functionals, #特徴量の計算レベルを決める。Functionals: LLDの統計値を計算する（デフォルト値）
)

feats = pd.DataFrame()  #dataframeの初期化
for file in files:
  feat = smile.process_file(file) #音声ファイルを入力とし、dataframe形式で値を返す
  feats = feats.append(feat)  #"feats ="がないと何も起こらない
```

```python
feats.to_csv("/content/features.csv")
```
